### Problema 1

```python
#Funcion que calcula las componentes de deformacion Ur, Uz
def Moggi(d,f,G,P,a):
    Ur=abs((3*(a**3)*P*d)/((4*G)*((f**2+d**2)**(1.5))))
    Uz=abs((3*(a**3)*P*f)/((4*G)*((f**2+d**2)**(1.5))))
    return Ur,Uz
```

```python
#Explicacion de los Parametros
#G=Modulo de Elasticidad (Pa)
#P=Exceso de presion producido por la fuente (Pa)
#a=radio de la fuente (m)
#f=Componente vertical de la fuente (m)
#d=Componente radial de la fuente (varia en cada caso para poder observar el perfil) (m)
```

Para este problema se transforma el efecto de deformacion a coordenadas cartesianas, obteniendo lo siguente:

De esta forma definimos una funcion que satisfaga estas nuevas ecuaciones.
Los Ejes (x,y), corresponden a la superficie de la corteza sujeta a deformacion.

```python
#Funcion que calcula las componentes de deformacion Ux,Uy,Uz
def Moggi(x,y,f,G,P,a):
    Ux=(3*(a**3)*P*x)/((4*G)*((f**2+d**2)**(1.5)))
    Uy=(3*(a**3)*P*y)/((4*G)*((f**2+d**2)**(1.5)))
    Uz=(3*(a**3)*P*f)/((4*G)*((f**2+d**2)**(1.5)))
    return Ux,Uy,Uz
```

Para comparar el vector proyectado y el no proyectado, asumiremos los sigiuintes parametros:

```python
#Parametros de la fuente Mogi
a=3*10**3 #m
P=20*10**6 #Pa
G=75*10**9 #Pa
f=9*10**3 #m
d=linspace(-300*10**3,300*10**3,100) #m
angle=30*(pi/180) #Radianes
```
